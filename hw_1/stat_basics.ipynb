### Задание 2. Работа с выбросами

В выбранных числовых столбцах найти выбросы, выдвинуть гипотезы об их причинах и проинтерпретировать результаты. Принять и обосновать решение о дальнейшей работе с ними.

```python
outliners_dict = {}
for i in [4, 5, 6, 16, 19, 20, 22]:
    print(horses.columns[i])
    median = horses[horses.columns[i]].median()
    print(f'Медиана у " {horses.columns[i]}" - {median}') 
    standart_deviation = horses[horses.columns[i]].std()
    quadratic_dispersion = horses[horses.columns[i]].std()*2
    print(f'Стандартное отклонение у "{horses.columns[i]}" - {standart_deviation:.2f}') 
    print(f'Квадрат стандартного отклонения у "{horses.columns[i]}" - {quadratic_dispersion:.2f}')  
    se = standart_deviation / (300**0.5) # стандартное отклонение средней = средняя / корень кол-ва измерений (корень 300)
    confidential_interval_95 = median + (1.96 * se)
    confidential_interval_99 = median + (2.58 * se)
    print(f'95% Доверительный интервал у "{horses.columns[i]}" - {confidential_interval_95:.2f}') 
    print(f'99% Доверительный интервал у "{horses.columns[i]}" - {confidential_interval_99:.2f}')
    
    q1 = horses[horses.columns[i]].quantile(0.25)
    q2 = horses[horses.columns[i]].quantile(0.5)
    q3 = horses[horses.columns[i]].quantile(0.75)
    iqr = q3 - q1 
    lower_outlier = q1 - 1.5 * iqr
    upper_outlier = q3 + 1.5 * iqr
    A={} #Создаем словарик из всех значений столбца, чтобы потом итерируясь по нему найти выбросы
    for j in horses.index:
        A[j] = horses[horses.columns[i]][j]
    #Находим выбросы
    outliers = {j:A[j] for j in A.keys() if (A[j] > upper_outlier) or (A[j] < lower_outlier)}
    print(f'Верхняя граница уса "{horses.columns[i]}" - {upper_outlier:.2f}')
    print(f'Нижняя граница уса "{horses.columns[i]}" - {lower_outlier:.2f}')  
    print(f'Выбросы у "{horses.columns[i]}" - {outliers}')
    print()
    outliners_dict[horses.columns[i]] = list(outliers.keys())
    #break
```

### pulse
При первом взгляде складывается ощущение, что выбросы за гранью верхнего уса и 99% доверительного интервала. Но если присмотреться к группам лошадей, разбитым по возрасту и переживанию боли, пульс сильно разнится:

```python
horses.groupby(['Age','pain']).pulse.median()
```

Посчитаем верхний ус:

```python
def calc_box_wiskers(data):
    q1 = data.pulse.quantile(0.25)
    q3 = data.pulse.quantile(0.75)
    iqr = q3 - q1 
    upper_outliner = q3 + 1.5 * iqr
    return upper_outliner
```

```python
horses.groupby(['Age','pain']).apply(calc_box_wiskers)
```

Выборка, которая попала в выброс: 

```python
horses.iloc[outliners_dict['pulse'] ,:][['Age', 'pain','pulse','surgery?']]
```

Не входит в объем задачи, но в категорийных качественных метриках, есть значения, которые не соответствуют описанию, например в значении метрики outcome всего три значения 1, 2, 3. В датафрейме значения 0.1 и 5.3, которые не подходят под описание.

```python
for name in horses.columns:
    if name in S: 
        
        for i in range(0,300,1):
            if horses.at[i,name] not in S[name]:
                horses.at[i,name] = 'NaN'
        
horses       
```
